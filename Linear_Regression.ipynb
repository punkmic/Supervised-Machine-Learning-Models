```python
import numpy as np
import matplotlib.pyplot as plt
import math, copy
```

### **Prepare data**

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}, y_train = {y_train}")
```

###  **Plotting the data**

```python
# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r')
# Set the title
plt.title('Housing Price')
# Set the y-axix label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.show()
```

### **Model Function**

```python
w = 200  # weight
b = 100  # bias
```

```python
def compute_model_output(x, w, b):
  m = x.shape[0] # number of examples
  f_wb = np.zeros(m) # init a new vector of m elements
  for i in range(m): # for each element
    f_wb[i] = w * x[i] + b # calculate the predict value (y) and add to the vector
  return f_wb # return our vector of predicted values (y)
```

### **Plot our predicted values**

```python
tmp_f_wb = compute_model_output(x_train, w, b,)

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
# Show legend
plt.legend()
plt.show()
```

### **Cost Fucntion**

```python
def compute_cost(x, y , w , b):
  m = x.shape[0] # number of examples
  cost_sum = 0
  for i in range(m):
    f_wb = x[i] * w + b # predicted value
    cost = (f_wb - y[i]) ** 2 # calculate the cost
    cost_sum += cost
  total_cost = cost_sum * (1/ (2 * m))
```

### **Compute the gradiente**

```python
def compute_gradient(x, y, w, b):
  m = x.shape[0]
  dj_dw = 0
  dj_db = 0

  for i in range(m):
    fb = w * x[i] + b
    dj_dw_i =(fb - y[i]) * x[i]
    dj_db_i = (fb - y[i])
    dj_db += dj_db_i
    dj_dw += dj_dw_i
  dj_dw = dj_dw / m
  dj_db = dj_db / m
  return dj_dw, dj_db 
```

### **Gradient Descent**

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iters, cost_function, grandient_function):
  w = copy.deepcopy(w_in)  # avoid modifying global w_in
  b = b_in
  w = w_in

  for i in range(num_iters):
    # Calculate the gradient and update the parameters using gradient_function
    dj_dw, dj_db = compute_gradient(x, y, w, b)

    # Update parameters 
    b = b - alpha * dj_db
    w = w - alpha * dj_dw
  return w, b
```

```python
# initialize parameters
w_init = 0
b_init = 0
# some gradient descent settings
iterations = 10000
tmp_alpha = 1.0e-2
# run gradient descent
w_final, b_final = gradient_descent(x_train ,y_train, w_init, b_init, tmp_alpha, 
                                                    iterations, compute_cost, compute_gradient)
print(f"(w,b) found by gradient descent: ({w_final:8.4f},{b_final:8.4f})")
```
